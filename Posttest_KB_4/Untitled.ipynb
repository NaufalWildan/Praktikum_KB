## Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

## Ubah nilai NULL yang memiliki tipe data desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["home_team_goalkeeper_score"] = imputer.fit_transform(df[["home_team_goalkeeper_score"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Ubah nilai NULL yang memiliki tipe data kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["home_team"] = imputer.fit_transform(df[["home_team"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

## Ubah nilai NULL yang memiliki tipe data bilang bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["away_team_mean_midfield_score"] = imputer.fit_transform(df[["away_team_mean_midfield_score"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```
