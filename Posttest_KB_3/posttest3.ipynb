## Library

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Membuka File

```python
df = pd.read_csv("international_matches.csv")
```

```python
df.head(10)
```

```python
df.info()
```

## Perbandingan Home team score dengan away team score


##### pada diagram di bawah menjelaskan perbandingan score yang di cetak tim saat berlaga sebagai home atau away, di mana score tim home lebih banyak di bandingkan score tim away

```python
plt.figure(figsize=(12,7))

x = df['away_team_score']
y = df['home_team_score']

plt.bar(x, y, color='lightcoral')

plt.title('Jumlah Perbandingan Score Team Home dan Away', size=16)
plt.ylabel('Home Team Score', size=14)
plt.xlabel('Away Team Score', size=14)
plt.xticks(size=12)
plt.yticks(size=12)

plt.show()
```

## korelasi 2 attribute dengan tipe data angka

```python
plt.figure(figsize=(10,5))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```

## Membuat diagram pie dengan attribut 

```python
data_pie = df['away_team_score'].value_counts()
```

```python
data_pie = df['away_team_score'].value_counts().reset_index(name = "score")
```

```python
plt.pie(data_pie.score,
    labels = data_pie["score"], 
    startangle=90, 
    autopct='%.1f%%', 
    shadow=True
)
plt.title('Diagram Pie Score team Away')
plt.show()
```
